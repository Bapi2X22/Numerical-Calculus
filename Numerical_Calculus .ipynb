# Numerical Differentiation

$\underline{Forward:} \hspace{1cm} \displaystyle{\lim_{h \to 0}} \dfrac{f(x_0+h)-f(x_0)}{h}$

```python
# =============================================================================
#   Forward
# =============================================================================

domain=[0,2*np.pi]
N=20
h=(domain[1]-domain[0])/(N-1)
   
x=np.linspace(domain[0],domain[1],N)
def f(x):
    return np.sin(x)

def forward_diff(x):
    return (f(x+h)-f(x))/h

# plt.figure(4,figsize=(8.5, 6.5))
# plt.plot(x,np.zeros(N),"-o")
# plt.plot(x,f(x),"ro")

plt.figure(5,figsize=(8.5, 6.5))
plt.plot(x,np.zeros(N),"-o")
plt.plot(x,forward_diff(x),"bo")
plt.plot(np.linspace(0,2*np.pi,100),np.cos(np.linspace(0,2*np.pi,100)))

```

# Backward Differentiation
$\underline{Procedure:} \hspace{1cm} \displaystyle{\lim_{h \to 0}} \dfrac{f(x_0)-f(x_0-h)}{h}$

```python
#########################  Backward  #######################################
domain=[0,2*np.pi]
N=20
h=(domain[1]-domain[0])/(N-1)
   
x=np.linspace(domain[0],domain[1],N)
def f(x):
    return np.sin(x)
def backward_diff(x):
    return (f(x)-f(x-h))/h

# plt.figure(6,figsize=(8.5, 6.5))
# plt.plot(x,np.zeros(N),"-o")
# plt.plot(x,f(x),"ro")

plt.figure(7,figsize=(8.5, 6.5))
plt.plot(x,np.zeros(N),"-o")
plt.plot(x[1:],backward_diff(x[1:]),"bo")
plt.plot(np.linspace(0,2*np.pi,100),np.cos(np.linspace(0,2*np.pi,100)))

```



# Central Differentiation
$\underline{Procedure:} \hspace{1cm} \displaystyle{\lim_{h \to 0}} \dfrac{f(x_0+h)-f(x_0-h)}{2h}$

```python
####################### Central #####################################
domain=[0,2*np.pi]
N=20
h=(domain[1]-domain[0])/(N-1)
   
x=np.linspace(domain[0],domain[1],N)
def f(x):
    return np.sin(x)

def centeral_diff(f,x):
    return (f(x+h)-f(x-h))/(2*h)

# plt.figure(8,figsize=(8.5, 6.5))
# plt.plot(x,np.zeros(N),"-o")
# plt.plot(x,f(x),"ro")

plt.figure(9,figsize=(8.5, 6.5))
plt.plot(x,np.zeros(N),"-o")
plt.plot(x[1:-1],centeral_diff(f,x[1:-1]),"bo")
plt.plot(np.linspace(0,2*np.pi,100),np.cos(np.linspace(0,2*np.pi,100)))

```

# Numerical Definite Integration Methods

$$\int_a^b f(x) dx \approx \sum_{i=1}^N f(x_i) W_i$$

1. Evenly spaced values of x between end points a and b
2. Number of points on x-axis = N
3. Number of intervals= N-1
4. Length of one interval = $h=\dfrac{b-a}{N-1}$

```python
# =============================================================================
# Integration
# =============================================================================

def f(x):
    return x**2*np.exp(-x)

plt.figure(14,figsize=(8.5, 6.5))
x=np.linspace(0,10,100)
plt.plot(x,f(x),"r")


domain=[0,10]
N=50                       # no need to take N>50
h=(domain[1]-domain[0])/(N-1)
   
x=np.linspace(domain[0],domain[1],N)

plt.plot(x,np.zeros(N),"-o")
plt.plot(x,f(x),"c-o")

```

# Trapezoidal Method

W= h$\big[\frac{1}{2}$, 1, 1, 1, 1 ......., 1, 1, $\frac{1}{2}\big]$

```python
W=[h]*N
W[0]=W[N-1]=h/2

trapezoidal_integration=np.sum(W*f(x))    ## W*f is element wise multiplication
print(trapezoidal_integration)
```

```python
######## Writing function #######################

def trapezoidal_int(f,domain,N):
    h=(domain[1]-domain[0])/(N-1)
    x=np.linspace(domain[0],domain[1],N)
    W=[h]*N
    W[0]=W[N-1]=h/2
    trapezoidal_int=np.sum(W*f(x)) 
    return trapezoidal_int

trapezoidal_int(f,[0,2*np.pi],50)
```

# Simpson Method

weights= $\frac{h}{3}\big[$ 1, 4, 2, 4, 2..........,2, 4, 1$\big]$

```python
def simpson(f,domain,N):
    h=(domain[1]-domain[0])/(N-1)
    x=np.linspace(domain[0],domain[1],N)
    W=[2*h/3 if (i%2==0) else 4*h/3 for i in range(N)]
    W[0]=W[N-1]=h/3
    integration=np.sum(W*f(x))
    return integration


simpson(f,domain,N)
```
